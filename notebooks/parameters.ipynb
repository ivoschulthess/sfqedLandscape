```python
import numpy as np
from scipy.constants import physical_constants
```

```python
# speed of light in [m/s]
clight = physical_constants['speed of light in vacuum'][0]

# reduced Planck constant
hbar_eVs = physical_constants['reduced Planck constant in eV s'][0]

# electron mass
me_MeV = physical_constants['electron mass energy equivalent in MeV'][0]
```

```python
##########
# INPUT  #
##########

# laser wavelength in [µm]
laserWavelength = 0.8

# laser intensity parameter
a0 = 0.83

# electron energy in [MeV]
E_e = 50

# collision angle in [deg]
theta = 3.7

##########
# OUTPUT #
##########

# laser angular frequency in [rad/s]
omega_l = 2*np.pi*clight / (laserWavelength * 1e-6)

# laser photon energy in [eV]
E_l = hbar_eVs * omega_l

# electron Lorentz factor
gamma_0 = E_e / me_MeV

# quantum parameter
chi = a0 * gamma_0 * E_l * (1 + np.cos(theta*np.pi/180)) / (1e6 * me_MeV)

# energy parameter
eta = chi / a0

print(f'a0:\t{a0:.6f}')
print(f'χ:\t{chi:.6f}')
print(f'η:\t{eta:.6f}')
```
